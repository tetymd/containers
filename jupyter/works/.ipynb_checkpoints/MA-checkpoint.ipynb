```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plot
import talib

%matplotlib inline
```

```python
# 株価データのインポート
df = pd.read_csv("kabuka.csv", encoding="shift-jis")
print(type(df))
df
```

```python
# usecol 使用するカラムを指定
df = pd.read_csv("kabuka.csv", encoding="shift-jis", header=1, index_col=0, usecols=[0, 4])
# 期間を指定
span = 25
```

```python
df.plot()
```

```python
# 単純移動平均をdfに追加
sma = df["終値"].rolling(window=span, min_periods=span).mean()
df["SMA25"] = sma
```

```python
df.plot()
```

```python
# 加重移動平均をdfに追加

b = np.ones(span) - 1
n = 0
for i in b:
    b[n] = b[n] + (n + 1)
    n += 1

def f(x):
    return np.sum(x * b) / (n * (n + 1) / 2)

wma = df["終値"].rolling(window=span, min_periods=span).apply(f, raw=False)
df["WMA25"] = wma
```

```python
df.plot()
```

```python
# 指数移動平均をdfに追加
close = np.array(df["終値"], dtype='f8')
close
```

```python
output = close.copy()
output = np.c_[output, talib.EMA(close)]
data = pd.DataFrame(output, index=df.index)[1]
df["EMA"] = data
```

```python
df.plot()
```
