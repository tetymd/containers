```python
# usecol 使用するカラムを指定
df = pd.read_csv("kabuka.csv", encoding="shift-jis", header=1, index_col=0, usecols=[0, 4])
df
```

```python
df.plot()
```

```python
# 単純移動平均をdfに追加
sma = df["終値"].rolling(window=25, min_periods=25).mean()
df["SMA25"] = sma
```

```python
df.plot()
```

```python
# 加重移動平均をdfに追加
wma = df.rolling(window=25, min_periods=25)

b = np.ones(25) - 1
n = 0
for i in b:
    b[n] = b[n] + (n + 1)
    n += 1

def f(x):
    return np.sum(x * b) / (n * (n + 1) / 2)

wma = df["終値"].rolling(window=25, min_periods=25).apply(f, raw=False)
df["WMA25"] = wma
df.plot()
```
